```python
import numpy as np
import pandas as pd
from textblob import TextBlob
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import SnowballStemmer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\amitk\Machine Learning\Twitter Sentiment Analysis Project\Book2.csv",encoding = 'ISO-8859-1')
df.head()
```

```python
def cleanTxt(text):
    text = re.sub(r'@[A-Za-z0-9]+','',text)     
    text = re.sub(r'#','',text)
    text = re.sub(r'RT[\s]+','',text)
    text = re.sub(r'https?:\/\/\S+','',text)
    
    return text

df["Text"] = df["Text"].apply(cleanTxt)
X = df["Text"]
X = X.values
print(X)
print(type(X))
```

```python
sw = set(stopwords.words('english'))
sw.remove('not')
sw.remove('nor')
```

```python
tokenizer = RegexpTokenizer('[a-zA-Z@.]+')
train_list = []
for sentence in X:
    train_list.append(tokenizer.tokenize(sentence.lower()) )

print(train_list)
```

```python
def remove_stopwords(text,stopwords):
    useful_words = [w for w in text if w not in stopwords]
    return useful_words

#send text as list of words.
final_list = []
for useful_tokens in train_list:
    final_list.append(remove_stopwords(useful_tokens,sw))
    
print(final_list)
```

```python
from nltk.stem import PorterStemmer
```

```python
clean_list = []

for useful_tokens in final_list:
    clean_list.append(' '.join(useful_tokens))

array = np.array(clean_list)
print(type(array))
B = np.reshape(array, (-1, 1))
print(B)
```

```python
sb = SnowballStemmer('english')
a = []

for sentence in clean_list:
    for word in sentence:
        a.append(sb.stem(word)) 

a = ' '.join(a)
print(a)
```
